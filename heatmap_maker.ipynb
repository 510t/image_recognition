```python
def heatmap_img(df):
    l_wid = np.floor(df[1] * 1600).astype(int)
    l_hei = np.floor(df[2] * 540).astype(int)
    r_wid = np.floor(df[3] * 1600).astype(int)
    r_hei = np.floor(df[4] * 540).astype(int)
    
    arr = np.zeros((1080, 1920, 1))

    for i in range(0, len(l_wid)):
        h = l_hei[i]
        w = l_wid[i]
        arr[h][w] += 1

    for i in range(0, len(r_wid)):
        h = r_hei[i]
        w = r_wid[i]
        arr[h][w] += 1

    img = array_to_img(arr).filter(ImageFilter.GaussianBlur(20))
    arr = img_to_array(img) / np.max(arr) * 255
    img = array_to_img(arr).filter(ImageFilter.GaussianBlur(20))
    arr = img_to_array(img) / np.max(arr) * 255
    img = array_to_img(arr).filter(ImageFilter.GaussianBlur(20))
    arr = img_to_array(img) / np.max(arr) * 255
    img = array_to_img(arr).filter(ImageFilter.GaussianBlur(20))
    arr = img_to_array(img) / np.max(arr) * 255
    
    return array_to_img(arr)
```

```python
def create_four_heatmap(filename):
    df = pd.read_csv('data/csv/' + filename + '.csv', header=None)
    first = df[:182].dropna(how='any').reset_index(drop=True)
    second = df[182:364].dropna(how='any').reset_index(drop=True)
    third = df[364:546].dropna(how='any').reset_index(drop=True)
    forth = df[546:].dropna(how='any').reset_index(drop=True)
    
    first_img = heatmap_img(first)
    second_img = heatmap_img(second)
    third_img = heatmap_img(third)
    forth_img = heatmap_img(forth)
    
    first_img.save('data/series/1/' + filename + '_heatmap.png')
    second_img.save('data/series/2/' + filename + '_heatmap.png')
    third_img.save('data/series/3/' + filename + '_heatmap.png')
    forth_img.save('data/series/4/' + filename + '_heatmap.png')
```

```python
import glob
x_paths = glob.glob('data/stimuli.out/*.jpg', recursive=False)

for path in x_paths:
    filename = path[17:-4]
    print(filename)
    create_four_heatmap(filename)
```
